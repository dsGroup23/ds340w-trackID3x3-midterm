```python
import os
import glob
import pandas as pd

IN_DIR  = "/Users/sunzhenhang/Desktop/TrackID3x3/output/CAMELTrack_outputs/Indoor/filtered_MOT"
OUT_DIR = "/Users/sunzhenhang/Desktop/TrackID3x3/output/CAMELTrack_outputs/Indoor/filtered_MOT_smooth"

os.makedirs(OUT_DIR, exist_ok=True)

mot_files = glob.glob(os.path.join(IN_DIR, "*.txt"))
print(f"Found {len(mot_files)} MOT files.")

for mot_file in mot_files:
    df = pd.read_csv(mot_file, header=None)
    df.columns = ["frame_id", "id", "x", "y", "w", "h",
                  "conf", "cls", "vis", "empty"]

    df = df.sort_values(["id", "frame_id"]).reset_index(drop=True)

    df[["x", "y"]] = df.groupby("id")[["x", "y"]].transform(
        lambda s: s.rolling(window=5, min_periods=1, center=True).mean()
    )

    track_lengths = df.groupby("id")["frame_id"].transform("count")
    df = df[track_lengths >= 10].copy()

    out_path = os.path.join(OUT_DIR, os.path.basename(mot_file))
    df.to_csv(out_path, header=False, index=False)
    print(f"Processed {os.path.basename(mot_file)} -> {out_path}")

print("All files processed. Smoothed results saved to", OUT_DIR)
```
